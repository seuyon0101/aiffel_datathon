```python
from haversine import haversine, Unit
```

```python
dfs['coord'] = list(zip(dfs['ìœ„ë„'], dfs['ê²½ë„']))
dfs['coord']

dfg['coord'] = list(zip(dfg['ìœ„ë„'], dfg['ê²½ë„']))
dfg['coord']
```

```python
import timeit
```

```python
def distance_count(x, dataset, criteria = 'ì»¤í”¼ì /ì¹´í˜') :
    cnt=-1
    df = dataset.copy()
#     start= timeit.default_timer()
    for i in df['coord'] :
        if haversine(i, x, unit= Unit.METERS) < 5 :
            cnt+=1
#     print(timeit.default_timer() - start)
    return cnt        
```

```python
tdfs =dfg.loc[(dfg['ì‹œêµ°êµ¬ëª…'] == 'ì„±ë‚¨ì‹œ') & (dfg['ìƒê¶Œì—…ì¢…ì¤‘ë¶„ë¥˜ëª…'] == 'ì»¤í”¼ì /ì¹´í˜')].copy()
```

```python
tdfs['near_5m'] = tdfs['coord'].apply(lambda x : distance_count(x, tdfs))
tdfs['near_5m'].groupby(tdfs['í–‰ì •ë™ëª…']).mean().sort_values(ascending=False)
```
