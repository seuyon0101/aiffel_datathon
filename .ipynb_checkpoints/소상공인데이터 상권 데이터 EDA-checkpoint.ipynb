## count within parameter

```python
from haversine import haversine, Unit
```

```python
dfs['coord'] = list(zip(dfs['위도'], dfs['경도']))
dfs['coord']

dfg['coord'] = list(zip(dfg['위도'], dfg['경도']))
dfg['coord']
```

```python
# dfg.loc[(dfg['시군구명'] == '성남시') & (dfg['상권업종중분류명'] == '커피점/카페')]
```

```python
import timeit
```

```python
def distance_count(x, dataset, criteria = '커피점/카페') :
    cnt=-1
    df = dataset.copy()
#     start= timeit.default_timer()
    for i in df['coord'] :
        if haversine(i, x, unit= Unit.METERS) < 5 :
            cnt+=1
#     print(timeit.default_timer() - start)
    return cnt        
```

```python
tdfs =dfg.loc[(dfg['시군구명'] == '성남시') & (dfg['상권업종중분류명'] == '커피점/카페')].copy()
```

```python
tdfs['near_5m'] = tdfs['coord'].apply(lambda x : distance_count(x, tdfs))
tdfs['near_5m'].groupby(tdfs['행정동명']).mean().sort_values(ascending=False)
```

```python
sns.scatterplot(data=tdfs[tdfs['행정동명'] =='정자1동'], y= '위도',x='경도', hue ='near_5m')
plt.show()
```

### 대분류별

```python
maincat = dfs['상권업종대분류명'].value_counts(normalize=True).multiply(100).round(1)
fig = plt.figure(figsize =(6,4))
ax = sns.barplot(x= maincat.index , y = maincat)
plt.title('상권업종대분류명')
ax.bar_label(ax.containers[0])
sns.despine()
plt.xticks(rotation = 75)
plt.show()
```

### 중분류별

```python
maincat = dfs['상권업종대분류명'].unique()
```

```python
fig = plt.figure(figsize =(15,12) )
for i,val in enumerate(maincat) :
    dff = dfs[dfs['상권업종대분류명'] == val]
    subcat = dff['상권업종중분류명'].value_counts(normalize=True).multiply(100).round(1).head()
    ax1 = fig.add_subplot(4,2,i+1)
    ax = sns.barplot(y= subcat.index , x = subcat)
    plt.title(f"{val} : 상권업종중분류명 ")
    ax.bar_label(ax.containers[0])
    sns.despine()
    plt.xlabel('count')
plt.tight_layout()
plt.show()
```
