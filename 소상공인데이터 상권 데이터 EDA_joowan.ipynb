```python
# 분기별 지역별 성장률 변동
fig = plt.figure(figsize =(15,30))
for i,val in enumerate(reg_mid[:9]) :
    ax1 = fig.add_subplot(9,3,i+1)
    grow_rate = []
    for j in range(len(comp_list)-1):
        temp_group = group_base + [comp_list[j]] 
        oc_cat_n_reg_group = surv_cat_n_reg.groupby(temp_group)['상가업소번호'].count()
        oc_cat_n_reg_group = oc_cat_n_reg_group.to_frame().reset_index()    
        open_quart = oc_cat_n_reg_group.loc[(oc_cat_n_reg_group['시군구명'] == val) & (oc_cat_n_reg_group[comp_list[j]] == 'open'),'상가업소번호'].sum()
        close_quart = oc_cat_n_reg_group.loc[(oc_cat_n_reg_group['시군구명'] == val) & (oc_cat_n_reg_group[comp_list[j]] == 'close'),'상가업소번호'].sum()
        dff_quart = open_quart - close_quart
        total_prev_store = surv_cat_n_reg.loc[surv_cat_n_reg['시군구명'] == val,quart_list[j]].sum()
        grow_rate.append(dff_quart / total_prev_store * 100)
        plt.plot(grow_rate)
    plt.title("{0} : 분기별 성장률, 평균 : {1:.2f}".format(val,np.mean(grow_rate)))
    plt.ylabel('percent') 
    plt.xticks(rotation = 75)
    ax1.set_xticks([0,1,2,3,4,5,6,7,8])
    ax1.set_xticklabels(comp_list)
    plt.tight_layout()
plt.show()
```
